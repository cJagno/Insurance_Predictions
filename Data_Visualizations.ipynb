```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/PT_Data_Science/PandasForDataManipulation/insurance.csv')
cf = pd.read_csv('/content/drive/MyDrive/PT_Data_Science/PandasForDataManipulation/insurance.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
cf['region'] = cf['region'].map({'southeast': 1,
                                   'southwest': 2,
                                   'northwest': 3,
                                   'northeast': 4})
```

```python
cf['sex'] = cf['sex'].map({'female': 1,
                           'male': 2})
```

```python
cf['smoker'] = cf['smoker'].map({'yes': 1,
                                 'no': 0})
```

```python
corr = cf.corr()
```

```python
sns.heatmap(corr, cmap='Blues', annot=True)
plt.title('Correlations');
```

```python
cff = cf[['age', 'bmi', 'charges', 'smoker']]
corr2 = cff.corr()
```

```python
sns.heatmap(corr2, cmap='autumn', annot=True)
plt.title('Correlations');
```

```python
sns.barplot(df['region'], df['charges'])
plt.title('Average Charges per Region')
```

```python
sns.barplot(df['sex'], df['charges'])
plt.title('Average Charges per Sex')
```

```python
smoker = df['smoker']=='yes'
nonsmoker = df['smoker']=='no'
```

```python
smoking_df = df.loc[smoker,:]
nonsmoking_df = df.loc[nonsmoker,:]
```

```python
sns.barplot(smoking_df['sex'], smoking_df['charges'])
plt.title('Average Insurance charges of Smokers')
```

```python
sns.barplot(nonsmoking_df['sex'], nonsmoking_df['charges'])
plt.title('Average Insurance charges of Non-Smokers')
```
